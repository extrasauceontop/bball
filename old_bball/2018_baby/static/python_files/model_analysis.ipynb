```python
import pandas as pd
import os
import numpy as np
```

```python
df = pd.read_csv("../../data/tested.csv")
df_anal = df[["outcomes_x", "preds"]]
df_anal["pred_outcome"] = df["preds"].round(0)
```

```python
df_anal["pred_absolute"] = df_anal["preds"]-0.5
df_anal["pred_absolute"] = df_anal["pred_absolute"].abs()
df_anal
```

```python
outcome_list = df_anal["outcomes_x"].to_list()
pred_list = df_anal["pred_outcome"].to_list()

corrects = []
counter = 0
for outcome in outcome_list:
    pred = pred_list[counter]
    
    if outcome == pred:
        correct = 1
    else:
        correct = 0
    
    corrects.append(correct)
    counter = counter+1

df_anal["correct"] = corrects
```

```python
bins = [0, 0.07999, 0.1599, 0.2399, 0.3199, 0.4499, 1]

# Create the names for the four bins
group_names = ["F", "D", "C", "B", "A", "S"]
```

```python
df_anal["pred_val_summary"] = pd.cut(df_anal["pred_absolute"], bins, labels=group_names, include_lowest=True)
df_anal
```

```python
group_df = df_anal.groupby("pred_val_summary")
group_df["correct"].count()
```
